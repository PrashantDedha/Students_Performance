#### How to change default aggregation in pivot table?

aggfunc

```python
pt4 = pd.pivot_table(data=df,index =['reth','lunch'],columns=['gender'],
                     values=['ws','ms'],aggfunc=['sum','min'])
pt4

# aggfunc = min, max, sum, mean, median, count etc
```

#### 2) set_index() and reset_index()
